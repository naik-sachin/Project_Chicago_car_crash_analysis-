Based on the first instance of our DTC model it seems there are lot of features that are barely contributing to the model (<1%) . Before we start to get rid of some of the features perhaps we should oversample the minority causes to avoid any class imbalance issues. 

```python
Crash_type_dicts = {}
keys = range(len(df.columns))
for i in keys:
        Crash_type_dicts[i] = df.columns[i]
```

```python
# summarize distribution
counter = Counter(y)
for k,v in counter.items():
    per = v / len(y) * 100
    print('n=%d %s, (%.3f%%)' % ( v,Crash_type_dicts[k], per))
# plot the distribution
# pyplot.bar(counter.keys(), counter.values())
# pyplot.show()
```

```python
#  oversampling minority class and undersampling majority class
smote = SMOTE()
X_train_resampled, y_train_resampled = smote.fit_sample(X_train, y_train) 
# summarize distribution
counter = Counter(y_train_resampled)
for k,v in counter.items():
    per = v / len(y_train_resampled) * 100
    print('n=%d %s, (%.3f%%)' % ( v,Crash_type_dicts[k], per))
# plot the distribution
# pyplot.bar(counter.keys(), counter.values())
# pyplot.show()
```

```python
X_rs_df = pd.DataFrame(X_train_resampled)
```

```python
# X_rs_df.info()
```

```python
DTC_model2 = DecisionTreeClassifier(criterion='entropy',random_state=42)
```

```python
DTC_model2.fit(X_train_resampled, y_train_resampled)
```

```python
y_test_pred = DTC_model2.predict(X_test)
```

```python
#model accuracy metric using precision and recall
print("F1 Score: {}".format(f1_score(y_test, y_test_pred, average= 'weighted')))
print(f'Accuracy Score Test: {accuracy_score(y_test, y_test_pred)}')
```

Seems accuracy and F1 went down so perhaps we should try using cross validation to make assure sampling is not an issue

```python
#We will use three repeats of 10-fold cross-validation,
#10-fold cross-validation is applied 5 times fitting and evaluating 50 models on the dataset.
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
scores = cross_val_score(DTC_model2,X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
#Check the Accuray score of the 30 models 
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

Cross validation on original training data barely improved . Lets try cross validating resampled smote data.

```python
#We will use three repeats of 10-fold cross-validation,
#10-fold cross-validation is applied 5 times fitting and evaluating 50 models on the dataset.
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
scores = cross_val_score(DTC_model2,X_train_resampled, y_train_resampled, scoring='accuracy', cv=cv, n_jobs=-1)
#Check the Accuray score of the 50 models 
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

Model accuracy went up significantly , which confirms our suspicion that class imbalance is having a significant impact on the ability of our model accuracy.

Lets look at all the featuresand see if they affect  our model. First lets look at the original training data

```python
# Using RFECV look at the cross-validated score performance for the original training data
DTC_model3 = DecisionTreeClassifier(criterion='entropy',random_state=42)
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv = RFECV(estimator=DTC_model3, step=1, cv=StratifiedKFold(n_splits=5),
              scoring='accuracy')
rfecv.fit(X_train, y_train)

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected with training data")
plt.ylabel("Cross validation score")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
# features to keep /remove based on training data and RFECV
for i in range(X_train.shape[1]):
    if rfecv.support_[i] ==False:
        print('Feature to remove: %s, Selected %s, Rank: %.3f' % (Crash_type_dicts[i], rfecv.support_[i], rfecv.ranking_[i]))
    
```

Based on training data it seems all except CRASH_TYPE_INJURY is important,Now lets look at how the features fare in the resampled smote data

```python

# Using RFECV look at the cross-validated score performance for the smote data
DTC_model4 = DecisionTreeClassifier(criterion='entropy',random_state=42)
# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv_s = RFECV(estimator=DTC_model4, step=1, cv=StratifiedKFold(n_splits=5),
              scoring='accuracy')
rfecv_s.fit(X_train_resampled, y_train_resampled)

print("Optimal number of features : %d" % rfecv.n_features_)

# Plot number of features VS. cross-validation scores
plt.figure()
plt.xlabel("Number of features selected with SMOTE data")
plt.ylabel("Cross validation score ")
plt.plot(range(1, len(rfecv_s.grid_scores_) + 1), rfecv_s.grid_scores_)
plt.show()
```

```python
# features to keep /remove based on resampled SMOTE data
for i in range(X_train_resampled.shape[1]):
    if rfecv_s.support_[i] ==False:
        print('Feature to remove: %s, Selected %s, Rank: %.3f' % (Crash_type_dicts[i], rfecv_s.support_[i], rfecv_s.ranking_[i]))
    
```

Looks like almost all the features were selected in both the resampled smote data and the original training data. Which means for the data we have at hand with the number of features and target classes, our model cant perform any better .Since the model accuracy is higher with smote data , Perhaps  to increase accuracy we need more data or reduce the number of target classes.Gathering more data is not possible at this point.So to remedy this that we can try reapeating stratified kfolds and RFE with cross validation as this repettitive model evaluating to some extent will mimic more data. For this we will use the smote Data as class imbalance is present in our data.


```python
# evaluate model using RFE for classification by using repeated stratified kfold
rfe = RFE(estimator=DecisionTreeClassifier())#half the features
DTC_model5 = DecisionTreeClassifier(criterion='entropy',random_state=42)
pipeline = Pipeline(steps=[('s',rfe),('m',DTC_model5)])
# evaluate model using 10 splits and repeat each 5 times
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
scores = cross_val_score(pipeline, X_train_resampled, y_train_resampled, scoring='accuracy', cv=cv, error_score='raise',n_jobs=-1)
# Accuracy
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
# features to keep /remove based on training data and RFECV
rfe.fit(X_train_resampled, y_train_resampled)
for i in range(X_train_resampled.shape[1]):
    if rfe.support_[i] ==True:
        print('Feature to keep: %s' % (Crash_type_dicts[i]))
    
```

```python
# evaluate model using RFECV for classification by using repeated stratified kfold
# create pipeline
rfecv = RFECV(estimator=DecisionTreeClassifier())#automatic feature selection using cross validation
DTC_model6 = DecisionTreeClassifier(criterion='entropy',random_state=42)
pipeline = Pipeline(steps=[('s',rfecv),('m',DTC_model6)])
# evaluate model using 10 splits and repeat each 5 times
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=42)
scores = cross_val_score(pipeline, X_train_resampled, y_train_resampled, scoring='accuracy', cv=cv,n_jobs=-1)
# report performance
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
# features to keep /remove based on resampled SMOTE data
rfecv.fit(X_train_resampled, y_train_resampled,cv=cv)
for i in range(X_train_resampled.shape[1]):
    if rfecv.support_[i] ==False:
        print('Feature to remove: %s, Selected %s, Rank: %.3f' % (Crash_type_dicts[i], rfecv.support_[i], rfecv.ranking_[i]))
    
```

We can see that automatically selecting features(all in our case) and then fitting a decision tree on the selected features gives us a classification accuracy close to 70 percent.

```python
rfe.fit(X_train_resampled, y_train_resampled)
rfe_features_to_keep =[]
rfe_features_to_remove=[]
# summarize all features

for i in range(X_train_resampled.shape[1]):
    print('Column: %d, Selected %s, Rank: %.3f' % (i, rfe.support_[i], rfe.ranking_[i]))
    if rfe.support_[i] == True:
        rfe_features_to_keep.append(i) 
    else :
        rfe_features_to_remove.append(i) 
```
